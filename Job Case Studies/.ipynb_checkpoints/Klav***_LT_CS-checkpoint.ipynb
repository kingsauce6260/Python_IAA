```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from lifetimes import GammaGammaFitter
from lifetimes.fitters.modified_beta_geo_fitter import ModifiedBetaGeoFitter
from lifetimes.generate_data import modified_beta_geometric_nbd_model
from lifetimes.plotting import plot_calibration_purchases_vs_holdout_purchases, plot_period_transactions
from lifetimes.utils import summary_data_from_transaction_data, calibration_and_holdout_data
```

# **Warmup/basic python:**
>`1. Implement the modified BG model from the lifetimes package using the data we provide.`

Before implementing the modified BG model(MBG), the data first needs to be in the correct format. To do so the `summary_data_from_transaction_data()` function was used. The variables created are as follows:<br>
**CustomerID:** ID of the customer<br>
**frequency:** The count of distinct dates that a customer has had a purchase on, minus one(amount of repurchases)<br>
**recency:** The age of the customer on the last purchase made<br>
**T:** The age of the customer on the last day of the data analyze<br>
**monetary_value:** Average value of past purchases

```python
# Data for the MBG model
mbg_summary = summary_data_from_transaction_data(data, 'CustomerID', 'Date', monetary_value_col='PurchaseValue', observation_period_end='2017-12-06')

print(mbg_summary.head())
```

```python
# Fit MBG
mbg = ModifiedBetaGeoFitter(penalizer_coef = 0.0)
mbg.fit(mbg_summary['frequency'], mbg_summary['recency'], mbg_summary['T'])
print(mbg)
```

>`2. List the 100 customers predicted to make the most purchases over the next 12 months.`

To predict the top 100 customers who purchase the most over the next 12 months a MBG model was used. More on the model will be discussed in the Assumptions section.

```python
# Setting time period to predict(days)
t = 365

# Creating predicted purchases for next 12 months
mbg_summary['predicted_purchases'] = mbg.conditional_expected_number_of_purchases_up_to_time(t, mbg_summary['frequency'], mbg_summary['recency'], mbg_summary['T'])

# Outputting customers with most purchases
most_purchases = mbg_summary.sort_values('predicted_purchases', ascending=False).head(100)
most_purchases = most_purchases.reset_index()
customers_predicted_most_purchases = most_purchases[['CustomerID', 'predicted_purchases']]
print(customers_predicted_most_purchases)
```

>`3. List the 100 customers predicted to spend the most over the next 12 months.`

To predict the top 100 customers to spend the most over the next 12 months a Gamma-Gamma model was used. More will be discussed in the Assumptions section.

```python
# Need the Gamma-Gamma model to produce a predicted spend
# Assumption for has to be repeat purchaser
returning = mbg_summary[mbg_summary['frequency'] > 0]

# Meeting monetary and frequency correlation assumption
print(returning[['monetary_value', 'frequency']].corr())

# Showing head of what our summary looks like with predicted_purchases
print(mbg_summary.head())
```

```python
# Fitting the Gamma-Gamma model
ggf = GammaGammaFitter(penalizer_coef = 0)
ggf.fit(returning['frequency'],
        returning['monetary_value'])
print(ggf)
```

```python
# Predicting spend
# Creating predicted spend
mbg_summary['Predicted_Spend'] = ggf.customer_lifetime_value(
    mbg, #the model to use to predict the number of future transactions
    mbg_summary['frequency'],
    mbg_summary['recency'],
    mbg_summary['T'],
    mbg_summary['monetary_value'],
    time=12, # in months
    discount_rate=0.01, # default
    freq='D'
)

# Creating list of top spending customers
top_spending = mbg_summary.sort_values('Predicted_Spend', ascending=False).head(100)
top_spending=top_spending.reset_index()
top_spending_customers = top_spending[['CustomerID', 'Predicted_Spend']]
print(top_spending_customers.head(5))
```

# **Analysis**
>`4. Explain the statistical assumptions that the model makes. Do you think those assumptions are valid for a model of a customer buying widgets from an ecommerce store?`

## MBG Assumptions
**Assumption 1:** Each individual, I, has a hidden $lambda_i$ and $p_i$ parameter.<br><br>
The first part means that each individual has a hidden rate of purchase/transaction rate, lambda(λ.) This assumes that each individual is different thus allowing for a distribution of λ’s instead of an unlikely constant λ. The second part of the assumption states that each individual has a unique dying probability of p, instead of an unlikely constant overall p.<br><br>
Does this make sense for an ecommerce widget store? Yes, each customer’s buying pattern, λ, is different and each has a different probability of death, p. It makes sense to have a distribution for shoppers’ transaction rates,  λ, rather than one λ for everyone, and it makes sense for there to be a distribution for shoppers’ probabilities of death, rather than having one probability for everyone.<br><br>
From here estimates of distributions can start to develop, that generate the parameters, essentially viewing customers’ behavior.
<br><br>

**Assumption 2:** The $lambda_i$ and $p_i$ parameters come from a population wide Gamma and a Beta Distribution respectively.<br><br>
This assumption states that $lambda_i$ comes from a Gamma Distribution of the individuals’ lambda population. In our case the Gamma Distribution, Gamma(r, α), helps predict wait time(T) for future events, utilizing the parameters: transaction rate(where r=λ; following a Poisson Distribution) and number of events you are waiting to occur(where k=α). It is important to note that a big attraction for the Gamma and Beta Distributions are that they have the ability to create significantly differing shapes(with only two parameters) providing flexibility.

The second part of this assumption states that $p_i$ comes from a Beta distribution of probabilities of dying. The Beta Distribution is known for its versatility in shape and can be used to adjust as new information comes in. In our case, MBG, Beta(a, b) creates a distribution of probabilities of dying.

Utilizing all parameters(α,r,a,b), we can start to learn and predict more about the customers. An example would be the probability of a customer being alive given their recency, frequency, and T.

Does this make sense for an ecommerce widget store? Yes, both distributions have high flexibility with minimal inputs. It also makes sense to have transaction rates and probabilities of dying be distributions instead of one overall constant number.
<br><br>

**Assumption 3:** Individuals purchases follow a Poisson process with rate $λ_i$*t.<br><br>
This assumptions helps predict the probability of events(purchases) happening, using the data about how often an event has occured in the past. Using lambda, λ, we are able to get probability for a given number of purchases for an individual in a fixed interval of time.

Does this make sense for an ecommerce widget store? Yes, Poisson Distributions are commonly used to forecast purchases on days/seasons. However, there is a concern that the λ is constant. As I discuss later, this may not be the best assumption if there are not  two full cycles of data available or if Black Friday is not modeled seperately. <br><br>

**Assumption 4:** At the beginning of their lifetime and after each purchase, an individual has a $p_i$ probability of dying (never buying again.)<br><br>
This assumption is where the MBG model differs from the BG model. Unlike BG, the MBG model allows customers to die after their first purchase, time(T)=0. In the BG model, individuals are not allowed to die until they have a repeat purchase. This is extremely important in the widget store case because of the extremely high concentration of one time purchasers, as you can see in the graph Number of Repeat Transactions Per Customer. This graph shows frequency of purchases ie number of repurchases.

```python
mbg_summary['frequency'].plot(kind='hist', bins=50, title='Number of Repeat Transactions Per Customer')
print(mbg_summary['frequency'].describe())
print(sum(mbg_summary['frequency'] == 0)/float(len(mbg_summary)))
```

After each point in time that a purchase occurs, the probability of death for the individual is reassessed. Hence, the point at which the individual dying is now distributed across transactions, using a shifted geometric distribution. The more repurchases there are the less uncertainty of the individual dying, the probability of dying decreases. This makes sense because when given a history of customer data you are going to likely base whether a customer will repurchase on their purchase history. You will generally have more confidence they will return if they continuously repurchase, hence uncertainty will go down.

Does this make sense for an ecommerce widget store? It mostly does. Yes, because maybe if the store has a ton of different widgets and the customer is loyal to the store your uncertainty of death will decrease if their business needs multiple widgets over time. It puts me more at ease that the assumption also includes that individuals can also die at the beginning of their lifetime, which aligns more with widget buyers. There is only so many widgets an individual can buy(usually.) Playing devil’s advocate, one could make the case, that in most cases, the more widgets an individual buys, the less likely they are to purchase again. This is because, again, how many widgets does a company/customer actually need?<br><br>

## Gamma-Gamma Assumptions
To predict the monetary value the Gamma-Gamma model is needed. The Gamma-Gamma model can predict the most likely value per transaction in the future. Combining our MBG and GG model, I am able to build a Customer Lifetime Value(CLV) for the next 12 months.

**Assumption 1:** The monetary value of a customer’s given transaction varies randomly around their average transaction value.<br><br>
This makes sense that the monetary output would include some sort of randomness and not a constant.

Does this make sense for an ecommerce widget store? Yes, if customers repurchase they will most likely hover around the mean of their average transaction value. This is especially incorporated into the widget store because widget costs do not have as a high variance compared to other product categories, being that they are all the same product, even though they may perform different tasks.<br><br>

**Assumption 2:** Average transaction values vary across customers but do not vary over time for any given individual.<br><br>
This makes sense that average transaction values will not vary over time for any given individual, because given that widgets are all under the same product category, customers may stay within a particular price limit.

An example to better explain my reasoning above. Person A does not mind spending money on food and will always pay the top price. However, Person A is very frugal when it comes to widget shopping, always looking for discounts and the best bang for buck. Compared to Person B, who is very frugal when it comes to food, always trying to be concious not to over spend. However, Person B does not mind spending a lot on widgets and wants the best for his/her website. People having different buying preferences for different product categories. Affirming that people  won't vary in average transaction values over time, given a specific product category.

Does this make sense for an ecommerce widget store? Yes, widgets are a certain category so I believe consumers will continue to spend their average amount over time.<br><br>


**Assumption 3:** The distribution of average transaction values across customers is independent of the transaction process.<br><br>
This assumption was checked earlier, before fitting the Gamma-Gamma model, and it was found that there was a correlation of 0.042791 between monetary and frequency values. 
<br><br>
**Assumption 4:**
The Gamma-Gamma model assumes only repeat purchases in the data, meaning a frequency>=1. This assumption was met when repeat purchases were filtered before fitting the Gamma-Gamma model.
<br><br>
**Assumption 5&6:**
The Gamma-Gamma model assumes the average transaction value, $z_i$, is distributed as such *$z_i$∼ gamma(p,ν)*, with *E($Z_i$|p,ν) = ζ = p/ν*. While also assuming the hyperparameter _ν_, has a distribution as such _ν ∼ gamma(q, γ)_.
<br><br>

# **Write a simulation**
>`5. Simulate how many customers are alive after 10 days, 1 year, 10 years and 100 years and how many purchases they have made in that time using the modified BG model. Use a random sample of 100 customers and show the results for 1 run of your simulation.`

Below the function, `simulation`, is used to simualte data for 100 random customers. From there the amount the amount alived can be summed up and the results are printed.

```python
# Create simulation function
def simulation(T, seed=67):
    """
    :summary: simulating data for 100 customers for specified time
    :param T: time range in days
    :return: frequency, recency, T, lambda, p, alive
    """
    output = modified_beta_geometric_nbd_model(T, r, alpha, a, b, size=100)
    
    return output
```

```python
# Assigning parameters
mbg_summary = mbg.summary
r = mbg_summary.iat[0,0]
alpha = mbg_summary.iat[1,0]
a = mbg_summary.iat[2,0]
b = mbg_summary.iat[3,0]

# Completing simulations 
day10 = simulation(T=10, seed=87)
year1 = simulation(T=365, seed=87)
year10 = simulation(T=365*10, seed=87)
year100 = simulation(T=365*100, seed=87)

# Counting the amount of people alive
day10_alive = day10.alive.sum()
year1_alive = year1.alive.sum()
year10_alive = year10.alive.sum()
year100_alive = year100.alive.sum()

# Printing out results
print(day10_alive)
print(year1_alive)
print(year10_alive)
print(year100_alive)
```

### Holdout
Another way to analyze how well the model performs is break the data into a calibration period dataset and a holdout dataset. This allows us to test on data not yet seen. To choose a calibration period, Jean-Rene Gauthier and Ben Van Dyke have run many simulates and they say that the typically calibration dataset should be five to ten times the typical inter purchase time of the customers. The minimum should be three times the typical inter purchase time of the customers, to get a good performance on the model. The test dataset length should be at least half the size of the training period, making sure the testing period is comparable to the training period. 

This means in our one year of data, after excluding the high counts at the end that are most likely Black Friday,  the minimum amount of months for the test dataset should be 4, half of 8. A discussion of why to have the end date 2017-11-27 follows.

```python
# Setting calibration date and end date
cal_date = data.Date.min() + datetime.timedelta(8 * 365 / 12)
cal_date = cal_date.strftime('%Y/%m/%d')
cal_date = pd.to_datetime(cal_date, format='%Y/%m/%d')
print(cal_date)
end_date = data.Date.max()
print(end_date)
```

```python
# Plotting the transaction count per date
time_data = pd.DataFrame()
time_data['date'] = data['Date']

time_data['true_time'] = time_data['date'].dt.strftime('%Y/%m/%d')
time_data['true_time'] = pd.to_datetime(time_data['true_time'], format='%Y/%m/%d')

time_data.sort_values(['true_time'], inplace=True)
time_data.rename(columns={'date':'transaction_count'}, inplace=True)
time_data.groupby(time_data['true_time']).count().plot(kind="line")

# Viewing top transaction counts 
count = time_data.groupby(time_data['true_time']).count()
print(count.sort_values('transaction_count').tail(10))
```

Because of these outliers in transaction count I chose to remove them. Statistically, if I wasn't aware of the outliers being around Black Friday, because it is not a full cycle I would not be able to confirm whether these outliers happen every year or if its a typo. Options are to impute, trim, or leave alone.

If there was a full cycle for both training and testing, meaning the data had two full years I would leave them in or model them seperately. Below a new end date is set.

```python
# Set new end date
end_date = data.Date.max()-datetime.timedelta(9)
print(end_date)
```

```python
# Creating calibration and holdout data
cal_hold = calibration_and_holdout_data(transactions=data, 
                                        customer_id_col='CustomerID', 
                                        datetime_col='Date',
                                        calibration_period_end=cal_date, 
                                        observation_period_end=end_date, 
                                        freq = 'D', 
                                        monetary_value_col='PurchaseValue')

# Fitting model
mbg = ModifiedBetaGeoFitter(penalizer_coef = 0.0000)
mbg.fit(cal_hold['frequency_cal'], cal_hold['recency_cal'], cal_hold['T_cal'])

plot_calibration_purchases_vs_holdout_purchases(mbg, cal_hold, n=10)
```

The plot above compares the average actual and predicted purchases. Notice that the prediction and the actuals are very close for customers with zero to four repeat transactions in the calibration period, while they diverge at five and come back closer at six. Another thing to be wary of is as you get further out there will be fewer customers.


Keeping the thinness of data for customers with higher repurchase values in mind, I plotted below a larger repurchase window. Looking at the plot, the model start to visably underperform for customers with higher repurchase values. This again may be due to just less data on these customers, but is a deficency on the model.

` 7. What are one or two ideas that could address the deficiencies of the model? (No need to implement, just describe.)`<br><br>
The biggest deficiencies of the model is the lack of ability to test a whole buying cycle, in this case a year, and how our model underperforms with customers who have more repurchases.A few recommendations:<br>
1) Collecting  two years of data to get 2 full cycles.<br>
2) Treating Black Friday and other anomolies seperately, ie maybe use a simulation model technique that models the extreme events such as Black Friday. Given enough data I would use an Exceedances technique with Generalized Pareto distribution if we wanted to get specific about those outlier dates.<br>
3) Once the model is confirmed to be good, both training and validation can be combined and forecasted out a desired period.<br>
4) Consider modeling for different coherts, ex. highly likely to purchase in the next year, instead of exact numbers.<br>
5) Join with a linear model with additional features or intervention variables.<br>

Suppose the intended use of the modified BG model is binning customers into one of the following categories: highly likely to purchase in the next year, somewhat likely to purchase in the next year, unlikely to purchase in the next year, highly unlikely to purchase in the next year. Suppose you want to compare this model with a model your colleague is proposing. How would you evaluate the two models? After binned into highly likely to purchase in the next year, somewhat likely to
purchase in the next year, unlikely to purchase in the next year, and highly unlikely to purchase in the next year, I would perform an ANOVA test to confirm that the models can be identified by their new segmentation bases, that were used during the segmentation process. This is important because I don’t want to be comparing models that binned the customers in the same way, this would give us nothing to compare.

Once segmented and depending on the type of store, compare values such as average spent per buying cycle, median products bought, quantity per order, how often they purchase, etc… By using ANOVA to statistically compare these model’s segments you are able to statistically confirm which is better at distinguishing these groups of customers. 

Here is a key to follow along with the example. Highly likely to purchase in the next year(Segment 1), somewhat likely to purchase in the next year(Segment 2), unlikely to purchase in the next year(Segment 3), and highly unlikely to purchase in the next year(Segment 4.)

An example: using average spent per buying cycle.

| Segment | Model 1 | Model 2 | Significance |
|:-------:|--------:|--------:|-------------:|
|    1    |     60  |     100 |      <0.0001 |
|    2    |     50  |      70 |      <0.0001 |
|    3    |     40  |      30 |      <0.0001 |
|    4    |     30  |      10 |      <0.0001 |

In this case with average spent per buying cycle as the variable analyzing the models, I would choose Model 2 as the better model because it better distinguishes the different types of segments.

If there were multiple metrics you want to compare, a voting method may be used to choose the better model or maybe a weighted voting system as well if some metric are more important than others.
